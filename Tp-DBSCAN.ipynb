```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Charger les donnes depuis multishapes.txt
data = pd.read_csv('/content/multishapes.txt', sep='\t')

# Slectionner les colonnes x et y pour clustering
X = data[['x', 'y']]

# Standardiser les donnes (important pour DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crer un modle DBSCAN
dbscan = DBSCAN(eps=0.15, min_samples=5)

# Appliquer le modle aux donnes
clusters = dbscan.fit_predict(X_scaled)

# Ajouter les clusters  la DataFrame
data['cluster'] = clusters

# Visualiser les rsultats
plt.scatter(data['x'], data['y'], c=data['cluster'], cmap='rainbow')
plt.xlabel('x')
plt.ylabel('y')
plt.title('DBSCAN Clustering')
plt.show()
```
