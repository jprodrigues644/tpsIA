```python
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt


def main():
    data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/barbecue.csv")
    print(data)
    print(data['barbecue'])

    x_train = data.drop('barbecue', axis=1)
    y_train = data['barbecue']

    model = tree.DecisionTreeClassifier(criterion='entropy')
    model.fit(x_train, y_train)

    plt.figure(figsize=(20, 10))  # Ajustez la taille selon vos besoins
    tree.export_graphviz(model, out_file='tree.dot',
                         feature_names=['Meteo', 'Amis', 'Vent', 'Jour'])
    tree.plot_tree(model, filled=True)
    plt.show()

# ...
if __name__ == '__main__':
    main()
```
